## A. Comprensión del Negocio (Business Understanding — CRISP-DM)

### A.1. Contexto y dominio del problema

#### Área de aplicación
<p style="text-align: justify; text-indent: 2em;">
Este proyecto propone analizar un dataset real de pacientes internados en UCIs de múltiples hospitales de EE. UU. para explorar si existe una relación cuantificable entre el IMC al momento de la admisión y la duración de la estadía en la unidad de cuidados intensivos. El análisis se enmarca en el dominio de la <strong>salud pública y la gestión hospitalaria basada en datos</strong>.

#### Relevancia del problema
<p style="text-align: justify; text-indent: 2em;">
Las Unidades de Cuidados Intensivos representan el eslabón más costoso y tecnológicamente intensivo del sistema de salud moderno. En este ecosistema, la duración de la estancia (Length of Stay, o LOS) es una métrica de desempeño hospitalario universal. Actúa como un proxy directo del consumo de recursos, los costos de internación y la morbilidad del paciente. Por consiguiente, la capacidad de identificar tempranamente qué perfiles de pacientes tienen una mayor propensión a requerir estancias prolongadas es una prioridad estratégica para la administración sanitaria y la planificación clínica.
<p style="text-align: justify; text-indent: 2em;">
Factores como la obesidad y el sobrepeso, medidos habitualmente a través del <strong>Índice de Masa Corporal (IMC)</strong>, han sido señalados en la literatura médica como potenciales predictores de complicaciones y mayor demanda de recursos hospitalarios. Sin embargo, la evidencia empírica al respecto no es concluyente y varía según la población estudiada.

#### Beneficiarios del análisis
<p style="text-align: justify; text-indent: 2em;">
Se beneficiarían de este análisis:

- Gestores hospitalarios: para anticipar la demanda de camas en UCI y optimizar la asignación de recursos.
- Equipos médicos de guardia y triage: para identificar perfiles de pacientes con mayor riesgo de estadías prolongadas.
- Investigadores en salud pública: para comprender el impacto de la obesidad en el uso de infraestructura crítica.
- Aseguradoras y financiadores de salud, que deben proyectar costos de internación en UCI.

### A.2. Formulación del problema
<p style="text-align: justify; text-indent: 2em;">
El problema central que aborda este proyecto es:

> Determinar si existe una asociación estadísticamente observable entre 
> el Índice de Masa Corporal (IMC) del paciente al momento de ingreso a 
> la UCI y la duración de su estancia en dicha unidad.
<p style="text-align: justify; text-indent: 2em;">
Esta pregunta es relevante porque:

1. El IMC es una variable sencilla de medir y registrar en la admisión hospitarlaria, por lo que podría servir como indicador temprano de riesgo.
2. La duración de la estancia en UCI (LOS, por sus siglas en inglés: *Length of Stay*) es una variable clave para la planificación y el costo hospitalario.
3. Una relación positiva o negativa entre ambas variables podría orientar decisiones médicas y administrativas desde el momento mismo del ingreso del paciente.
<p style="text-align: justify; text-indent: 2em;">
El análisis se centrará en pacientes adultos con registros válidos de peso y altura al momento de la admisión, considerando también variables contextuales como sexo, edad, tipo de UCI y diagnóstico de admisión.
